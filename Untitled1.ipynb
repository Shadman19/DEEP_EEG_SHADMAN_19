```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
import keras
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
bclfr = Sequential()
```

```python
bclfr.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu', input_dim = x_train.shape[1])) #input_dim=640
```

```python
bclfr.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
bclfr.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
bclfr.compile(optimizer = 'SGD', loss = 'binary_crossentropy', metrics = ['accuracy']) # optimizer = 'adam'
```

```python
bclfr.fit(x_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = bclfr.predict(x_test)
```

```python
y_pred
```

```python
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```
